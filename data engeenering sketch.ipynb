```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# List of all pages i want to combine
urls = [
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2025-26",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2024-25",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2023-24",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2022-23",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2021-22",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2020-21",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2019-20",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2018-19",
    "https://clarksonathletics.com/sports/mens-ice-hockey/stats/2017-18"
]

all_seasons = []

for url in urls:
    print(f"Fetching {url} ...")
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    # Season label from URL
    season = url.split("/")[-1]

    # Try to find *team* table
    team_table = soup.find("table")
    if team_table:
        df_team = pd.read_html(str(team_table))[0]
        df_team["Season"] = season
        df_team["Type"] = "Team"
        all_seasons.append(df_team)

    # Extract *all other tables* (players, etc.)
    tables = soup.find_all("table")[1:]
    for idx, t in enumerate(tables):
        df = pd.read_html(str(t))[0]
        df["Season"] = season
        df["Type"] = f"Table_{idx+1}"
        all_seasons.append(df)

# Combine
print("Combining all data ...")
combined = pd.concat(all_seasons, ignore_index=True)

# Write to CSV
output_csv = "clarkson_mhockey_all_seasons.csv"
combined.to_csv(output_csv, index=False)

print(f"Done! Saved combined file to {output_csv}")
```

```python
import pandas as pd

# -----------------------------
# 1. LOAD FILE
# -----------------------------
df = pd.read_csv("clarkson_mhockey_all_seasons.csv")

# Keep player rows only
player_col = [c for c in df.columns if "Player" in str(c)][0]
df = df[df[player_col].notna()].copy()
df = df.rename(columns={player_col: "Player"})

# -----------------------------
# 2. FIND GP COLUMN
# -----------------------------
def find_gp_column(columns):
    for c in columns:
        name = str(c).lower().replace(" ", "")
        if name in ["gp", "games"]:
            return c
    for c in columns:
        if "gp" in str(c).lower():
            return c
    return None

gp_col = find_gp_column(df.columns)
df = df.rename(columns={gp_col: "GP"})
df = df[df["GP"] > 0]

# -----------------------------
# 3. SEPARATE SKATERS & GOALIES
# -----------------------------
if "GAA" in df.columns:
    goalies = df[df["GAA"].notna()].copy()
    skaters = df[df["GAA"].isna()].copy()
else:
    skaters = df.copy()
    goalies = pd.DataFrame()

# -----------------------------
# 4. PER-GAME NORMALIZATION
# -----------------------------
stats = ["G", "A", "SOG"]
for stat in stats:
    if stat in skaters.columns:
        skaters[f"{stat}_per_GP"] = skaters[stat] / skaters["GP"]

# -----------------------------
# 5. CREATE PTS SAFELY
# -----------------------------
if "PTS" not in skaters.columns:
    if "G" in skaters.columns and "A" in skaters.columns:
        skaters["PTS"] = skaters["G"] + skaters["A"]

skaters["PTS_per_GP"] = skaters["PTS"] / skaters["GP"]

# -----------------------------
# 6. EFFICIENCY METRICS
# -----------------------------
if "G" in skaters.columns and "SOG" in skaters.columns:
    skaters["Shot_Efficiency"] = skaters["G"] / skaters["SOG"]

# -----------------------------
# 7. CAREER TOTALS
# -----------------------------
career_totals = (
    skaters
    .groupby("Player", as_index=False)
    .sum(numeric_only=True)
)

# -----------------------------
# 8. TEAM STATS BY SEASON (SAFE)
# -----------------------------
agg_cols = ["GP", "G", "A", "PTS", "SOG"]
agg_cols = [c for c in agg_cols if c in skaters.columns]

team_by_season = (
    skaters
    .groupby("Season", as_index=False)[agg_cols]
    .sum()
)

# -----------------------------
# 9. EFFICIENCY INDEX
# -----------------------------
efficiency_index = skaters[
    ["Player", "Season", "GP",
     "G_per_GP", "A_per_GP", "PTS_per_GP",
     "SOG_per_GP", "Shot_Efficiency"]
].copy()

# -----------------------------
# 10. EXPORT
# -----------------------------
output_file = "clarkson_hockey_capstone_ready.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    skaters.to_excel(writer, sheet_name="Skaters_Seasonal", index=False)
    if not goalies.empty:
        goalies.to_excel(writer, sheet_name="Goalies_Seasonal", index=False)
    career_totals.to_excel(writer, sheet_name="Career_Totals", index=False)
    team_by_season.to_excel(writer, sheet_name="Team_By_Season", index=False)
    efficiency_index.to_excel(writer, sheet_name="Efficiency_Index", index=False)

print("SUCCESS — Capstone-ready dataset created:", output_file)
```

```python
import numpy as np
import pandas as pd
import re

def split_penalties(val):
    # Handle missing or non-string values
    if pd.isna(val):
        return pd.Series([np.nan, np.nan])
    
    val = str(val).strip()
    
    # If it contains a dash like "5-10"
    if "-" in val:
        parts = val.split("-")
        
        # If exactly two numbers, perfect
        if len(parts) == 2:
            try:
                return pd.Series([int(parts[0]), int(parts[1])])
            except:
                return pd.Series([np.nan, np.nan])
        
        # If more than two parts, take the first two
        if len(parts) > 2:
            try:
                return pd.Series([int(parts[0]), int(parts[1])])
            except:
                return pd.Series([np.nan, np.nan])
    
    # If it's a single number like "5"
    if val.isdigit():
        return pd.Series([int(val), np.nan])
    
    # If nothing matches
    return pd.Series([np.nan, np.nan])
```

## Remove non-player summary rows

The raw export includes rows that summarize the whole team or opponents.  
Those rows are removed so the final dataset contains only player records.


```python
import pandas as pd

# Load raw CSV
df = pd.read_csv("clarkson_mhockey_all_seasons.csv")

# Keep ONLY Table_1
master = df[df["('Type', '')"] == "Table_1"].copy()

# Remove summary rows
summary_rows = ["Team Team", "Total", "Opponents"]
master = master[~master["('Player', 'Player')"].isin(summary_rows)].copy()

# Rename columns
rename_map = {
    "('#', '#')": "jersey_no",
    "('Player', 'Player')": "player",
    "('GP', 'GP')": "gp",
    "('Shots', 'G')": "shots_g",
    "('Shots', 'A')": "shots_a",
    "('Shots', 'PTS')": "shots_pts",
    "('Shots', 'SH')": "shots_sh",
    "('Shots', 'SH%')": "shots_sh_pct",
    "('Shots', '+/-')": "plus_minus",
    "('Goals', 'PPG')": "goals_ppg",
    "('Goals', 'SHG')": "goals_shg",
    "('Goals', 'FG')": "goals_fg",
    "('Goals', 'GWG')": "goals_gwg",
    "('Goals', 'GTG')": "goals_gtg",
    "('Goals', 'OTG')": "goals_otg",
    "('Goals', 'HTG')": "goals_htg",
    "('Goals', 'UAG')": "goals_uag",
    "('Penalties', 'PN-PIM')": "penalties_pn_pim",
    "('Penalties', 'MIN')": "penalties_min",
    "('Penalties', 'MAJ')": "penalties_maj",
    "('Penalties', 'OTH')": "penalties_oth",
    "('BLK', 'BLK')": "blocks",
    "('Bio Link', 'Bio Link')": "bio_link",
    "('Season', '')": "season",
    "('Type', '')": "type",
}

master = master[list(rename_map.keys())].copy()
master = master.rename(columns=rename_map)

#  CORRECT NAME FIX (SAFE — NO CUTTING)
def fix_duplicate_names(name):
    if pd.isna(name):
        return name
    
    name = str(name).strip()
    words = name.split()
    
    half = len(words) // 2
    if words[:half] == words[half:]:
        return " ".join(words[:half])
    
    return name

master["player"] = master["player"].apply(fix_duplicate_names)

# Preserve table source
master["table_type"] = master["type"]

# Drop duplicates
master = master.drop_duplicates()

# Export
master.to_excel("clarkson_multiseason_master_TABLE1_ONLY.xlsx", index=False)

print(master.shape)
print(master["player"].head(10))
```
